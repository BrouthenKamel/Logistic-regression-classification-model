```python
x = dataset.drop(columns="target")
x
```

```python
y = dataset['target']
y
```

```python
def sigmoid(x):
  return 1 / ( 1 + np.exp(-x) )
```

```python
def model(datapoint, w, b):
  return sigmoid( np.dot(datapoint, w) + b )
```

```python
def loss(datapoint, y , w , b):
  return ( -(y) * np.log(model(datapoint, w, b)) - (1-y) * np.log(1- model(datapoint,w,b)) )
```

```python
def cost(x, y, w, b):
  total = x.shape[0]
  c = 0
  for i in range(total):
    c += loss(x[i,:], y[i], w, b)
  return c / total
```

```python
def gradient_descent(x,y, w,b):
  m,n = x.shape
  dj_db = 0
  for i in range(m):
    dj_db += model(x[i,:], w, b) - y[i]
  dj_db = dj_db / m
  dj_dw = np.zeros(n)
  for j in range(n):
    for i in range(m):
      dj_dw[j] += ( model(x[i,:], w, b) - y[i] ) * x[i,j]
    dj_dw[j] = dj_dw[j] / m
  return dj_dw, dj_db
```

```python
def update_params(x,y, w,b, alpha):
  m,n = x.shape
  dj_dw, dj_db = gradient_descent(x,y,w,b)
  new_w = np.zeros(n)
  new_b = 0
  for j in range(n):
    new_w[j] = w[j] - alpha * dj_dw[j]
  new_b = b - alpha * dj_db
  return new_w, new_b
```

```python
x_train = x.to_numpy()
print(x_train.shape)
y_train = y.to_numpy()
print(y_train.shape)
w = np.zeros(11)
b = 0
alpha = 0.00008
iterations = 2000
```

```python
def predict(x, threshold):
  if x>threshold:
    return 1
  else:
    return 0
```

```python
m,n = x_train.shape
correct = 0
for i in range(iterations):
  if (i % 10 == 0):
    print("iteration : {} --> cost = {}".format(i, cost(x_train, y_train, w, b)))
  w,b = update_params(x_train, y_train, w,b, alpha)
for i in range(m):
  if ( predict( model( x_train[i,:] , w, b) , 0.5) == y_train[i] ) :
    correct += 1
print("accuracy = {}".format(( correct / m ) * 100) )
```
